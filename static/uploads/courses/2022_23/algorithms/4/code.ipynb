# Naive Bubblesort
Implement here the naive bubblesort version, run the algorithm with all the lists above and see how long it takes to sort them.

```python
def naive_bubblesort(A : list) -> Tuple[list, int]:
    """ sort the input array A on place, using the Bubblesort algorithm """
        
    iteration = 0
    
    for _ in range(len(A)):
        
        for j in range(len(A)-1):
            
            if A[j] > A[j+1]: 
                
                a = A[j]
                
                A[j] = A[j+1]
                
                A[j+1] = a
                
            iteration += 1
                
    return A, iteration
```

### After the algorithm implementation measure how long it takes to sort the lists running the cell below

```python
tic = time.time()
naive_bubblesort(A=ascending_order_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")

tic = time.time()
res = naive_bubblesort(A=random_shuffle_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")

tic = time.time()
naive_bubblesort(A=descending_order_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")
```

# Bubblesort, an easy optimization

The naive bubblesort does $n^2$ operations.<br/> ***We can do better!*** A very simple optimization could be using as a starting point for the second loop the current index of the first loop. <br/>In this way we reduce the amount of operation!<br/> ***Can you guess what is the reduction factor?***

```python
def bubblesort(A : list) -> Tuple[list, int]:
    """ sort the input array A on place, using the Bubblesort algorithm """
    
    iteration = 0
        
    for i in range(len(A)):
        
        for j in range(0, len(A) - i - 1):
            
            if A[j] > A[j + 1]: 
                
                a = A[j]
                
                A[j] = A[j + 1]
                
                A[j + 1] = a
                
            iteration += 1
            
    return A, iteration
```

```python
tic = time.time()
bubblesort(A=ascending_order_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")

tic = time.time()
res = bubblesort(A=random_shuffle_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")

tic = time.time()
bubblesort(A=descending_order_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")
```

# Bubblesort, a further optimization!
Now it is possible to improve the Bubblesort using a ***boolean flag***. <br/>Write the code of improved Bubblesort and see the computational time changes.

```python
def flag_bubblesort(A : list) -> Tuple[list, int]:
    """ sort the input array A on place, using the Bubblesort algorithm """
    
    iteration = 0
    
    for i in range(len(A)):
        
        swap_flag = False
        
        for j in range(len(A) - i - 1):
            
            if A[j] > A[j + 1]: 
                
                a = A[j]
                
                A[j] = A[j + 1]
                
                A[j + 1] = a
                
                swap_flag = True
                
        
        iteration += 1

        if not swap_flag:
                        
            return A, iteration
        
    return A, iteration
```

```python
tic = time.time()
flag_bubblesort(A=ascending_order_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")

tic = time.time()
flag_bubblesort(A=random_shuffle_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")

tic = time.time()
flag_bubblesort(A=descending_order_list.copy())
toc = time.time()

print(f"Time needed to order the list {toc-tic}")
```
