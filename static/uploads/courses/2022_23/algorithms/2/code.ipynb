## Fibonacci, a first recursive approach...
Recursion is a very elegant way to solve problems. Anyway, sometime it is highly inefficient...<br />
First of all we start importing the python library **time** that will be useful to measure the amount of time needed to get the results.

````python
import time
from functools import wraps

# You can just run this cell ignoring everything inside!

def timeit(func):
    """
    Function decorator, it is used to get the function's computational time
    If applied to a recursive function can be used to get the computational time
    for each call.
    
    To use the decorator you have to add on top of the funciton definition
    @timeit and then declare the function.
    If you want disable the decorator you can just delete @timeit
    """
    
    @wraps(func)
    def timeit_wrapper(*args, **kwargs):
        """
        Function that effectively compute the function's time
        """
        
        tic = time.time()
        result = func(*args, **kwargs)
        toc = time.time()
        
        print(f'Function {func.__name__}{args} {kwargs} Took {toc-tic:.7f} seconds to compute {result}')
        return result
    
    return timeit_wrapper
````

## Coding styles
 
 
There are two coding styles: **Object-Oriented** (explicit) and **PyPlot-Style** (implicit). The differences are mainly about objects references. In OO-style we act on figures and axes otherwise we use the plt reference. 

## Example

### OO-Style

```
ax.plot(x, x, label='linear') 
ax.plot(x, x**2, label='quadratic') 
ax.plot(x, x**3, label='cubic')

```

### PyPlot-Style

```
plt.plot(x, x, label='linear') 
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

```


## The simplest plot you can draw

````python
import matplotlib.pyplot as plt
data = [1, 2, 3, 4]
plt.plot(data)
plt.show()
````

The statement makes available matplotlib API

```
import matplotlib.pyplot as plt

```

the *data* list instead is just a bunch of numbers we want to plot

```
data = [1, 2, 3, 4]

```

The **plt.plot()** function takes as input the values to plot (and other parameters that we will see later). **plt.show()** display all open figures.

```
plt.plot(data)
plt.show()

```
**Note**: *if you provide a single list or array to plot, matplotlib assumes it is a sequence of y values,
and automatically generates the x values for you. Since python ranges start with 0, the default x vector 
has the same length as y but starts with 0*



## Figures and Axes
Matplotlib is organized in figures and axes, this makes it flexible enough to draw figures with more than a single plot in it. Each axis (a subplot) is independent from the others and can be custumized as you prefer

````python
import numpy as np

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

fig_2, (ax1_2, ax2_2) = plt.subplots(nrows=2, ncols=1, figsize=(4, 10))


fig.suptitle("A figure with 2 axes", fontsize=18)
fig_2.suptitle("Another figure with 2 axes", fontsize=18)


data_series_1 = np.random.rand(10)
data_series_2 = np.random.rand(10)

ax1.plot(data_series_1)
ax2.plot(data_series_2)

ax1.set(title="data series 1", xlabel="x unit", ylabel="y unit" )
ax2.set(title="data series 2",  xlabel="x unit", ylabel="y unit" )

ax1_2.plot(data_series_1)
ax2_2.plot(data_series_2)

ax1_2.set(title="data series 1", xlabel="x unit", ylabel="y unit" )
ax2_2.set(title="data series 2",  xlabel="x unit", ylabel="y unit" )

plt.show()
````

It could be the case that we have many different axes at the same time in the same figure, for example 16 different subplots. In order to deal with them correctly we must use a for loop to initialize them. First of all again we build the figure and then we iterate over all the axes using

```
fig_3.axes

```

Then we plot and set properties for each subplot with 

```
axis.plot()
axis.set()


```

````python
fig_3, _ = plt.subplots(nrows=4, ncols=4, figsize=(10, 10))

for index, axis in enumerate(fig_3.axes):
    axis.plot(np.random.rand(10))
    axis.set(title=f"data series {index}", xlabel="x unit", ylabel="y unit" )
    
plt.tight_layout()
plt.show()
````

Matplotlib enable us to personalize each subplot independently, for instance we can use a dictionary to set different properties of each subplots

````python
properties = {
    "plot_1":
    {
        "title": "TSLA stock price",
        "xlabel": "Days",
        "ylabel": "Dollars ($)",
    },
    
    "plot_2":
    {
        "title":  "TSLA volumes",
        "xlabel": "Days",
        "ylabel": "Dollars ($)",
    },
    
    "plot_3":
    {
        "title":  "TSLA 21 period MA",
        "xlabel": "Days",
        "ylabel": "Dollars ($)",
    },
    
    "plot_4":
    {
        "title":  "TSLA 7 period MA",
        "xlabel": "Days",
        "ylabel": "Dollars ($)",
    }
}

fig_3, _ = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

for index, (axis, (key, props)) in enumerate(zip(fig_3.axes, properties.items())):
    axis.plot(np.random.rand(10))
    axis.set(title=props["title"], xlabel=props["xlabel"], ylabel=props["ylabel"])
    
plt.tight_layout()
plt.show()
````

## Multiple data series

It is possible to plot multiple data series within a single axis calling ``` ax.plot() ``` multiple times also using different plot types as you can see below (plot, scatter, errorbar).<br/>
**Now try to modify something in the code below, try to understand how things work**

````python
import random

fig_4, ax_4 = plt.subplots(nrows=1, ncols=1, figsize=(6,6))

x = [0, 1, 2, 3, 4, 5, 6]

x_squared = [element**2 for element in x]

x_2_5 =  [element**2.5 for element in x]

random_noise = [random.uniform(2, 5) for _ in range(len(x_squared))]

random_error = [random.randrange(2, 4) for _ in range(len(x_squared))]

ax_4.plot(x_squared, label="$x^2$")

ax_4.scatter(x, np.add(x_squared, random_noise), c="red", label="$x^2$ + noise")

ax_4.errorbar(x, x_2_5, yerr=random_error, c="orange", label="$x^{2.5}$ + errors")

ax_4.tick_params(axis='both', which='major', labelsize=18)

ax_4.set_title(label="Just a bunch of data series", fontsize=20)

ax_4.set_xlabel(xlabel="Time (second)", fontsize=18)

# Using 'r' and '$' before and within a string in matplotlib allows you to use LaTeX sintax!
ax_4.set_ylabel(ylabel=r"Speed $(\frac{m}{s})$", fontsize=18)

ax_4.legend(prop={'size': 15})

plt.tight_layout()
````

## Best Practices

There are a serie of best practices to follow in order to get pretty plots!!

- Use large enough font size for axes labels and major ticks labels 
- Use always a legend, it should be readable so use a large font size
- Avoid overlapping between plots and legend
- Use always y and x labels and if needed insert units of measure!!! 
- Use for the major tick label a font size large enough to be read easily but remember to leave enough space between two labels
